```python
import tkinter as tk
from deep_translator import GoogleTranslator        
# Deep translator is flexible free and unlimited python tool to translate between different languages in a 
# simple way using multiple translators. It is capable of bulk translations that is a file from one language to another.
# It has support for multiple translation APIs, including Google, Microsoft, Yandex, Libre, and others.

window = tk.Tk()                                     # Create window
window.title("Language Translator with GUI")         # It gives the name to the window

def translate_lang():

    from_lan = "en"
    to_lan = "hi"
    
    if input_lang.get() == 'English':
        from_lan = 'en'
    elif input_lang.get() == 'Hindi':
        from_lan = 'hi'
    elif input_lang.get() == 'Spanish':
        from_lan = 'es'
    elif input_lang.get() == 'German':
        from_lan = 'de'
    elif input_lang.get() == 'French':
        from_lan = 'fr'
        
    if output_lang.get() == 'English':
        to_lan = 'en'
    elif output_lang.get() == 'Hindi':
        to_lan = 'hi'
    elif output_lang.get() == 'Spanish':
        to_lan = 'es'
    elif output_lang.get() == 'German':
        to_lan = 'de'
    elif output_lang.get() == 'French':
        to_lan = 'fr'
    
    translation = GoogleTranslator(source=from_lan, target=to_lan).translate(Text.get())  
    # It can auto detect language using source = 'auto'
    # Here we give the text as arguement that is to be translated from one language to another
    output.set(translation)                          # Translation is the translated text displayed as output
    
input_lang = tk.StringVar()                          # declaring string variable
output_lang = tk.StringVar()

Language_choices = ['Hindi', 'English', 'French', 'German', 'Spanish']
input_lang.set('English')                            # set the value of the input_lang variable
output_lang.set('Hindi')

tk.Label(text="Choose a language").grid(row = 0, column = 1)    # Label() is widget for displaying text on window
input_lang_menu = tk.OptionMenu(window, input_lang, *Language_choices)  
# OptionMenu() is a widget used to display the options in dropdown manner
input_lang_menu.grid(row = 1, column = 1)                      # It displays widget in grid like manner

tk.Label(text="Translated language").grid(row = 0, column = 3)
output_lang_menu = tk.OptionMenu(window,output_lang, *Language_choices)
output_lang_menu.grid(row = 1, column = 3)

tk.Label(text = "Enter Text: ").grid(row = 2, column = 0)

Text = tk.StringVar()
Textbox = tk.Entry(textvariable = Text)
Textbox.grid(row = 2, column = 1)

tk.Label(text = "Translated Text: ").grid(row = 2, column = 2)

output = tk.StringVar()
Textbox = tk.Entry(textvariable = output)
Textbox.grid(row = 2, column = 3)

translate_btn = tk.Button(text = 'Translate', bd = 5, relief = tk.RAISED, command = translate_lang)
translate_btn.grid(row = 3, column = 2)

window.mainloop()
```
