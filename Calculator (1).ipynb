```python
import tkinter as tk
window = tk.Tk()               # Create window
window.title("Calculator")     # It gives the name to the window

def click(number):
    global operation
    operation += str(number)
    input_text.set(operation)  # set the value of the input variable
    
def clear():
    global operation
    operation = ""
    input_text.set(operation)
    
def calculate():
    global operation
    
    if operation[-2:] == "/0":
        input_text.set("Not possible")
        operation = ""
    else:  
        output = str(eval(operation)) # eval() is an inbuilt python function which takes a string argument(equation) 
                                      #  It returns a string(output of the equation).
        input_text.set(output)
        operation = output
    
    
operation=""                      # Declaring operation variable to store the expression that is to be calculated

input_text = tk.StringVar()       # declaring string variable
input_ent = tk.Entry(font = ('large, _font', 15, 'bold'), bd=10, justify = 'right', insertwidth=4, textvariable = input_text)
# insertwidth -> By default the insertion cursor is 2 pixels wide. You can adjust this by setting insertwidth to any dimension.
# bd -> The size of the border around the indicator. Default is 2 pixels. 
# textvariable -> In order to be able to retrieve the current text from your entry widget using the created StringVar variable.
input_ent.grid(columnspan=10)  # columnspan specifies How many columns widget occupies by default it's value is 1.
 
btn7 = tk.Button(font = ('arial',15,'bold'), text='7', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(7))
# Lamda Functions (also referred to as Anonymous Function in Python) are very useful in building Tkinter GUI applications. 
# They allow us to send multiple data through callback function. In Button Command, It is used to pass data to callback function
# Lambda can be inside any function that works as an anonymous function for expressions. 
btn7.grid(row = 1, column = 0)

btn8 = tk.Button(font = ('arial',15,'bold'), text='8', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(8))
btn8.grid(row = 1, column = 1)

btn9 = tk.Button(font = ('arial',15,'bold'), text='9', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(9))
btn9.grid(row = 1, column = 2)

btn_sum = tk.Button(font = ('arial',15,'bold'), text='+', bg='tan', bd=5, padx= 13, pady = 10, command = lambda:click('+'))
btn_sum.grid(row = 1, column = 3)

btn4 = tk.Button(font = ('arial',15,'bold'), text='4', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(4))
btn4.grid(row = 2, column = 0)

btn5 = tk.Button(font = ('arial',15,'bold'), text='5', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(5))
btn5.grid(row = 2, column = 1)

btn6 = tk.Button(font = ('arial',15,'bold'), text='6', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(6))
btn6.grid(row = 2, column = 2)

btn_sub = tk.Button(font = ('arial',15,'bold'), text='-', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click('-'))
btn_sub.grid(row = 2, column = 3)

btn1 = tk.Button(font = ('arial',15,'bold'), text='1', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(1))
btn1.grid(row = 3, column = 0)

btn2 = tk.Button(font = ('arial',15,'bold'), text='2', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(2))
btn2.grid(row = 3, column = 1)

btn3 = tk.Button(font = ('arial',15,'bold'), text='3', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(3))
btn3.grid(row = 3, column = 2)

btn_mul = tk.Button(font = ('arial',15,'bold'), text='*', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click('*'))
btn_mul.grid(row = 3, column = 3)

btn0 = tk.Button(font = ('arial',15,'bold'), text='0', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click(0))
btn0.grid(row = 4, column = 0)

btn_clear = tk.Button(font = ('arial',15,'bold'), text='C', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:clear())
btn_clear.grid(row = 4, column = 1)

btn_equal = tk.Button(font = ('arial',15,'bold'), text='=', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:calculate())
btn_equal.grid(row = 4, column = 2)

btn_div = tk.Button(font = ('arial',15,'bold'), text='/', bg='tan', bd=5, padx= 15, pady = 10, command = lambda:click('/'))
btn_div.grid(row = 4, column = 3)

window.mainloop()           # Continuously display the window in loop
```
